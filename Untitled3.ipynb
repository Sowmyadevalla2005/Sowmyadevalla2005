```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(data.iloc[:,:-1].astype(float).corr(),linewidths=0.1,vmax=1.0,
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' has the correct columns 'meanfun' and 'meanfreq'
g = sns.FacetGrid(data, hue="label", height=5)
g.map(sns.scatterplot, "meanfun", "meanfreq")
g.add_legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Assuming `data` is your DataFrame, already defined
X = data.iloc[:, :-1]
y = data['label']
y = y.replace({'male': 1, 'female': 0})

# Rescale data before TSNE
X_scale = StandardScaler().fit_transform(X)

# Sklearn TSNE
tsne = TSNE(n_components=2, init='pca', random_state=0)
X_t = tsne.fit_transform(X_scale)

# Plot
plt.figure()
plt.scatter(X_t[y == 0, 0],  # Corrected indexing for males (y == 0)
            X_t[y == 0, 1],
            marker='x', color='g',
            linewidth=1, alpha=0.8, label='Male')
plt.scatter(X_t[y == 1, 0],  # Corrected indexing for females (y == 1)
            X_t[y == 1, 1],
            marker='v', color='r',
            linewidth=1, alpha=0.8, label='Female')

plt.xlabel('Dim 1')
plt.ylabel('Dim 2')
plt.title('T-SNE')
plt.legend(loc='best')
plt.show()
```
